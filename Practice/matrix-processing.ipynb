```python
#Add Row to Matrix)
matrix = [
    [1, 2],
    [3, 4]
]
for row in matrix:
    print(row)
```

```python
# Add new row
matrix.append([5, 6])
for row in matrix: 
    print(row)
```

```python
#Add Multiple Rows
matrix.extend([[9, 6], [7, 8]]) 
for row in matrix:
    print(row)

```

```python
############# Mathematical Operations on Matrix
A = [
    [1, 2],
    [3, 4]
]

B = [
    [5, 6],
    [7, 8]
]
print("Matrix A:")
for row in A:
    print(row)  
print("\nMatrix B:")
for row in B:
    print(row)
```

```python
# len() is a function provided by Python that returns the number of items in an object.
#  In the case of a matrix (which is a list of lists),
#  len(A) gives the number of rows, and 
# len(A[0]) gives the number of columns (assuming all rows have the same number of columns). 

print("\nNumber of rows in A:", len(A))
print("Number of columns in A:", len(A[0]))
print("\nNumber of rows in B:", len(B))
print("Number of columns in B:", len(B[0]))
```

```python
print(len("Hello"))
print(len([1, 2, 3, 4, 5]))
print(len({'a': 1, 'b': 2, 'c': 3}))    
```

```python
# adding two matrices
result = []

for i in range(len(A)): # len() is used to get the number of rows in the matrix
    row = []
    for j in range(len(A[0])):
        row.append(A[i][j] + B[i][j])
    result.append(row)

print(result)
```

```python
# Matrix Subtraction
#Matrix A:
#[1, 2]
#[3, 4]

#Matrix B:
#[5, 6]
#[7, 8]
result = []

for i in range(len(A)):
    row = []
    for j in range(len(A[0])):
        row.append(A[i][j] - B[i][j])
    result.append(row)

print(result)
```

```python
# Multiplication of two matrices
result = [] #Not presized hence sum_product is initialized to 0 for each element in the result matrix
for i in range(len(A)):
    row = []
    for j in range(len(B[0])):
        sum_product = 0 # Initialize sum_product to 0 for each element in the result matrix
        for k in range(len(B)):
            sum_product += A[i][k] * B[k][j]
        row.append(sum_product)
    result.append(row)
print(result)
```

```python
result = [[0, 0], [0, 0]] # Presize the result matrix with zeros

for i in range(len(A)):
    for j in range(len(B[0])):
        for k in range(len(B)):
            result[i][j] += A[i][k] * B[k][j]

print(result)
```

```python
# transpose of a matrix
transpose = list(zip(*A)) # zip(*A) is used to unpack the rows of A and zip them together, effectively transposing the matrix.
print(transpose)
# zip() can be used to combine two or more iterables (like lists) element-wise. When used with the unpacking operator (*), it can transpose a matrix by zipping together the rows of the original matrix.
# zip(* ) is a common Python idiom for transposing a matrix. The * operator unpacks the rows of the matrix A, and zip() then groups the first elements of each row together, the second elements together, and so on, effectively creating the transposed version of the matrix.
```

```python
# division of two matrices is not defined in the traditional sense, but we can perform element-wise division if the matrices are of the same size. Here's how you can do it:
result = []
for i in range(len(A)):
    row = []
    for j in range(len(A[0])):
        if B[i][j] != 0: # Check to avoid division by zero
            row.append(A[i][j] / B[i][j])
        else:
            row.append(float('inf')) # or you could choose to append None or some other value to indicate division by zero
    result.append(row)
print(result)
```
