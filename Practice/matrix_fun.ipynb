```python
# A matrix is just a table of numbers arranged in rows and columns.
# In Python, we represent it using a list of lists or libraries like NumPy.

# Think of it like an Excel sheet—but handled using code.
```

```python
row1 = [1, 2, 3]
row2 = [4, 5, 6]

matrix = [row1, row2]
print(matrix)
print(matrix[0])  # Accessing the first row
print(matrix[0][1])  # Accessing the
# element in the first row, second column
for row in matrix:
    print(row)
```

```python
header = ["Name", "Age", "City"]
data = [
    ["Alice", 30, "New York"],
    ["Bob", 25, "Los Angeles"],
    ["Charlie", 35, "Chicago new york rendering"],
]
table = [header, *data]
print(table)    
for row in table:
    print(f"{row[0]:<10} {row[1]:<5} {row[2]:<15}")
```

```python
print("Name :", table[1][0])  # Accessing Alice's name
print("Age :", table[1][1])   # Accessing Alice's age   
print("City :", table[1][2])  # Accessing Alice's city
```

```python
print("header of the table :", table[0])  # Accessing the header of the table
```

```python
# append a new row to the table
# extend the table with a new row
# extend the list data with a new row
data.append(["David", 28, "Miami"])
data.append(["Eve", 22, "Seattle"])
data.extend([["shruti", 60, "India"], ["Evey", 22, "Seattle"]])
table = [header, *data]
   
for row in table:
    print(f"{row[0]:<10} {row[1]:<5} {row[2]:<15}")
```

```python
# Creating a matrix using list of lists
sales = [
    [100, 200, 150],
    [80, 220, 130],
    [90, 210, 160]
]
print("Sales Matrix:")
for row in sales:
    print(row)
```

```python
rows = 2
cols = 3

matrix = []

for i in range(rows):
    row = []
    for j in range(cols):
        value = int(input(f"Enter value for ({i},{j}): "))
        row.append(value)
    matrix.append(row)

print("Matrix:")
for r in matrix:
    print(r)
```

```python
rows = 1
cols = 4   # Name and Class, Roll No, branch


# Add header
matrix = [["Name", "Class", "Roll No", "Branch"]]
while True:
    for i in range(rows):
        row = []
    
        name = input(f"Enter Name for person : ")
        class_name = input(f"Enter Class for person : ")
        roll_no = input(f"Enter Roll No for person : ")
        branch = input(f"Enter Branch for person : ")

# print ID card for each person
    print("\nID Card:")
    print
    print(f"Name: {name}")
    print(f"Class: {class_name}")
    print(f"Roll No: {roll_no}")
    print(f"Branch: {branch}") 
    print("-" * 20)

    row.append(name)
    row.append(class_name)
    row.append(roll_no)
    row.append(branch)

    matrix.append(row)

    print("\nMatrix:")
    for r in matrix:
     print(r)

# do you want to add more people to the matrix?
    cont = input("Do you want to add more people? (yes/no): ")
    if cont.lower() != "yes":
        break

```

```python
print("\nFormatted Output:\n") #\n = new line (line break)

for r in matrix:
    print(f"{r[0]:<10} {r[1]:<5}") # f" = formatted string, :<10 = left-align with width 10 spaces, :<5 = left-align with width 5 spaces
```
