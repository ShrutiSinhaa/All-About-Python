```python
# What is Boolean?

#   A Boolean is a data type that has only two possible values:

#👉 True
#👉 False

#That’s it! Nothing in between.

#You can think of it like a switch:

#ON → True
#OFF → False
```

```python
# we use capital T and F:

a = True
b = False

print(a)
print(b)
```

```python
# Using Boolean in Decision Making
age = int(input("Enter your age: "))

if age >= 18: # this is boolean expression
    print("You can vote") # you can add a else statement to handle the case when age is less than 18
else:
    print("You cannot vote yet")
```

```python
status = input("Are you enrolled (True/False): ").lower()
# The .lower() method converts the input to lowercase, 
# so it can handle "True", "true", "False", and "false" inputs.
# it would also handle "TRUE" and "FALSE" inputs, as they will also be converted to lowercase.

enrolled = status == "true"

if enrolled:
    print("Student is enrolled")
else:
    print("Student is not enrolled")
```
