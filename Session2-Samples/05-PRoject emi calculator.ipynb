```python
# Project - data types and operations
# Check loan eligibility based on income
# Calculate the EMi
# Print the Message based on the eligibility
```

```python
# INPUT
Name = input("Enter your name: ")
income = float(input("Enter your monthly income: "))
loan_amount = float(input("Enter loan amount: "))
time = int(input("Enter time (years): "))
rate = 5
# LOGICAL CHECK (Eligibility)
if income >= 30000:
    
    # EMI CALCULATION
    monthly_rate = rate / (12 * 100)
    months = time * 12
    emi = (loan_amount * monthly_rate * (1 + monthly_rate) ** months) / ((1 + monthly_rate) ** months - 1)
    print("Dear", Name)
    print("\nCongratulations! You are eligible for the loan ✅")
    print("\nYour approved load details are as follows:")
    print("\nLoan Amount: ", loan_amount)
    print("Interest Rate: ", rate)
    print("Time Period: ", time, "years")
    print("Your EMI is:", round(emi, 2))

else:
    print("Sorry", Name)
    print("\nYou are not eligible for the loan ❌")
```
