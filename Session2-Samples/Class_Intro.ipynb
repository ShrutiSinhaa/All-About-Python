```python
A class is a blueprint (template) to create objects.

Real-life analogy:

Class = Blueprint of a house
Object = Actual house built from that blueprint

Why class : 
Data + behavior are grouped together
Code becomes clean, reusable, scalable
```

```python
class Student:
    def __init__(self, name, marks): # __init__ is a special method called constructor, 
        # it initializes the object with given values
        # self is a reference to the current instance of the class,
        #  it allows us to access the attributes and methods of the class
        self.name = name
        self.marks = marks

    def display(self): # Functions inside class are called methods. always use self as the first parameter in methods to access instance variables
        print("Name:", self.name)
        print("Marks:", self.marks)

s1 = Student("John", 85) # s1 is an object (instance) of the Student class, created using the constructor __init__
s2 = Student("Alice", 90)
s1.display()# calling the display method of the s1 object to show its details
s2.display()
```
