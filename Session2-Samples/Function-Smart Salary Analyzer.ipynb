```python
PROJECT: Smart Salary Analyzer

A program that:

Takes employee name and salary
Calculates tax
Calculates net salary
Gives a message based on salary range
```

```python
# Smart Salary Analyzer using Functions

# 1. Input Function
def get_employee_details():
    name = input("Enter employee name: ")
    salary = float(input("Enter salary: "))
    return name, salary

```

```python

# 2. Tax Calculation Function
def calculate_tax(salary):
    if salary > 50000:
        tax = salary * 0.20
    else:
        tax = salary * 0.10
    return tax

```

```python

# 3. Net Salary Function
def calculate_net_salary(salary, tax):
    net_salary = salary - tax
    return net_salary

```

```python

# 4. Salary Category Function
def salary_category(salary):
    if salary > 70000:
        return "Grade 3"
    elif salary >= 40000:
        return "Grade 2"
    else:
        return "Grade 1"

```

```python

# 5. Display Function
def display_result(name, salary, tax, net_salary, category):
    print("\n------ Salary Report ------")
    print(f"Employee Name : {name}")
    print(f"Salary        : {salary}")
    print(f"Tax           : {tax}")
    print(f"Net Salary    : {net_salary}")
    print(f"Category      : {category}")
    print("---------------------------")
```

```python


#  Main Program Flow
def main():
    name, salary = get_employee_details()
    tax = calculate_tax(salary)
    net_salary = calculate_net_salary(salary, tax)
    category = salary_category(salary)

    display_result(name, salary, tax, net_salary, category)

```

```python

# Run Program
main()
```
