```python
# 1. What is a Function?

A function is a reusable block of code.

👉 Instead of writing the same code again and again, you:

define it once
call it whenever needed
```

```python
def greet(): #defining a function named greet, it would define the code but not execute it until we call it
    print("Hello, welcome!")
```

```python
greet()  # calling the function to execute its code
```

```python
Parameter	Variable in function definition
Argument	Value passed when calling the function
```

```python
def display(name):   # parameter
    print("Hello", name)

display("shruti")    # argument
```

```python
name = input("Enter your name: ")#taking input from user and storing it in a variable named name
age = int(input("Enter your age: "))#taking input from user and storing it in a variable named age, converting it to integer
def display(name,age):   # parameter
    print("Hello, my name is", name, "and I am", age, "years old.")

display(name,age)    # argument
```

```python
# Passing default values to parameters:
# If user doesn’t pass a value → default is used

def greet(name="Guest"):
    print("Hello", name)

greet()          # Hello Guest
greet("Dev")  # Hello Dev
```

```python
# Returning values from functions:
# Functions can return values using the return statement. 
# This allows you to capture the output of a function and use it elsewhere in your code.

def square(x):
    return x * x

result = square(4)
print(result)
```

```python
# Scope: Local vs Global Variables

def test():
    x = 10
    print(x)

test(5) # This will raise an error because x is a local variable and not defined outside the function

test() # This will work because x is defined within the function and can be accessed when the function is called
```

```python
# global variable
x = 10
def test(x):
    print(x)

test(x) # This will work because x is a global variable and can be accessed inside the function
test(5) # This will work because we are passing a value to the function which will override the global variable x within the function's scope
```

```python
def add(a, b):
    """This function adds two numbers""" # docstring to explain what the function does 
    return a + b

result = add(3, 5)
print(result)  # Output: 8
```
