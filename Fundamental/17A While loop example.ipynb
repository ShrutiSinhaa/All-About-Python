```python
#Use while when:

#   You don’t know how many times to repeat
#   You depend on a condition
```

```python
x = 1
while x <= 10:
    print(x)
    x= x+1 # it would never end if we forget to increment x as it would be always 1 and the condition would always be true. 
    # This is called an infinite loop.
```

```python
x = "y"

while x == "y":
    number = int(input("Enter a number: "))
    
    i = 1
    while i <= 10:
        print(number, "X", i, "=", number * i)
        

```
