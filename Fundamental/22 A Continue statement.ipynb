```python
while True:
    table_number = int(input("Enter a number to print its multiplication table(Type 0 to exit`): "))
    if table_number == 0:
        break
    if table_number < 0:
        print("Please enter a positive number.")
        continue

    i = 1
    while i <= 10:
        print(f"{table_number} x {i} = {table_number * i}")
        i += 1
```
