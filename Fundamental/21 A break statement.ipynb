```python
while True:
    number = int(input("Enter a number to print its table: "))
    i = 1
    while i <= 10:
        print(f"{number} x {i} = {number * i}")
        i += 1
    
    choice = input("Do you want to print another table? (yes/no): ")
    if choice.lower() != "yes":
        break
```
